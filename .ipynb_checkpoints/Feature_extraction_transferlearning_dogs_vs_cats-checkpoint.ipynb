```python
import tensorflow as tf
from tensorflow import keras
```

```python
from keras.layers import Conv2D,Dense,MaxPool2D,BatchNormalization,Flatten,Rescaling
from keras.models import Sequential
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_ds=keras.utils.image_dataset_from_directory(directory='/content/dogs_vs_cats/train',labels='inferred',label_mode='int')
valid_ds=keras.utils.image_dataset_from_directory(directory='/content/dogs_vs_cats/test',labels='inferred',label_mode='int')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 10))
for i, (image, label) in enumerate(train_ds.take(9)):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(image[i].numpy().astype('uint8'))
    plt.title(int(label[i]))
    plt.axis("off")
```

```python
from keras.layers import Rescaling,Resizing,RandomRotation,RandomZoom,RandomContrast,RandomFlip,RandomTranslation,Input
```

```python
image_batch, label_batch = next(iter(train_ds))
```

```python
image=image_batch[0].numpy().astype('uint8')
```

```python
plt.imshow(image)
```

```python
data_aug=Sequential([RandomFlip(),RandomRotation(factor=0.2),RandomZoom(0.3),RandomContrast(0.2),RandomTranslation(0.2,0.2)])
```

```python
plt.figure(figsize=(13,13))
for i in range(9):
  plt.subplot(3,3,i+1)
  plt.imshow(data_aug(image).numpy().astype('uint8'))
```

```python
from keras.applications.vgg16 import VGG16
```

```python
conv_base=VGG16(include_top=False,weights='imagenet',input_shape=(256,256,3))
```

```python
conv_base.summary()
```

```python
model=Sequential()
```

```python
model.add(Rescaling(1.0/255,input_shape=(256,256,3)))
```

```python
model.add(RandomContrast(0.2))
model.add(RandomRotation(0.2))
model.add(RandomZoom(0.2))
model.add(RandomTranslation(0,0.2))
model.add(RandomFlip('Horizontal'))
```

```python
model.add(conv_base)
```

```python
model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
conv_base.trainable=False
```

```python
model.summary()
```

```python
history=model.fit(train_ds,validation_data=valid_ds,epochs=10)
```

```python
plt.plot(history.history['accuracy'],label='train')
plt.plot(history.history['val_accuracy'],label='validation')
plt.legend()
plt.show()
```

```python
image=image.reshape(1,256,256,3)
model.predict(image)
```
